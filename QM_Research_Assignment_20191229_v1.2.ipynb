Define tables and variables of interest from Eurostat based on eurostat table of content found [here (pdf download)](http://ec.europa.eu/eurostat/estat-navtree-portlet-prod/BulkDownloadListing?sort=1&file=table_of_contents_en.pdf).
```python
#===(2.2) DEFINE LIST OF TABLES OF INTEREST

euro_tables = ["urb_cpop1","urb_cpopcb"]
```

```python
#===(2.3) DATA DOWNLOAD FUNCTION
def f_euro_api_data_extract(p_table_list):
    
    all_dataframes = {}
    
    #for each table, (i) download data (ii) convert to dataframe (iii) save as .csv
    for table in p_table_list:
    
        #(i) extract dataset
        print("\n")
        print(table.upper(),"downloading ...")
        print("========================================================\n")
        dataset = client.get_dataset(table)
        print(dataset.label)

        #(ii) convert to pandas data frame
        dataframe = dataset.to_dataframe()
        print(dataframe.head())
        
        #no filtering of data on call
        
        #(iii) save as .csv
        time_now = datetime.datetime.now().strftime('%Y%m%d_%H%M%S')
        dataframe.to_csv(f"0_Data_download/{table}_{time_now}.csv")
        
        all_dataframes[table] = dataframe
        
        print(table.upper(),"csv file saved.")
    print("\n========================================================")
    print("All tables downloaded.")
    
    return all_dataframes
```

```python
#===(2.4) CALL DATA DOWNLOAD FUNCTION (uncomment to run)

run_locally = True

if run_locally is True:
    print("API not run, will attempt to use local files")
else:
    eurostat_api_dataframes = f_euro_api_data_extract(euro_tables)
```

```python
#===(3.1) REPORT COUNT ALL UNIQUE VARIABLE

#store unique variables
unique_variable_list = []

if api_use_working is True:
    df_dictionary = eurostat_api_dataframes
    print("Using API files\n")
else:
    if local_use_working is True:
        df_dictionary = table_all
        print("Using local files\n")
    else:
        print("No files to use")

#cycle through unique keys in dictionary
for key in df_dictionary:
    
    for variable in df_dictionary[key].indic_ur.unique():
        unique_variable_list.append(variable)
    print(f"Unique variables in {key}: {len(df_dictionary[key].indic_ur.unique())}")

print(f"Unique variables across dataframes: {len(unique_variable_list)}")
```

```python
#===(3.2) WRITE OUT DESIRED VARIABLES AND CHECK THEY EXIST

euro_selected_variables = ["DE1001V","DE2003V","DE2002V","DE2001V","DE2003I","DE2002I","DE2001I"]
euro_variable_present = 0
euro_variable_absent = 0


for variable in euro_selected_variables:
    if variable in unique_variable_list:
        print(f"===[ALL GOOD]=== Desired variable ({variable}) present")
        euro_variable_present += 1
    else:
        print(f"===[WARNING]==== Desired variable ({variable}) absent")
        euro_variable_absent += 1
```

```python
#===(3.4) TRANSPOSE

#break multi index
eu_df_filter_by_var_break = pd.DataFrame(eu_df_filter_by_var.to_records())

#drop unwanted columns
eu_df_filter_by_var_break.drop(columns = ["level_0","level_1"],inplace = True)

#set new index (leaving only variable column and values)
eu_df_filter_by_var_index = eu_df_filter_by_var_break.set_index(keys = ["cities","time","indic_ur"], drop = True)

#pivot table by unstacking one level
eu_df_filter_by_var_index_unstack = eu_df_filter_by_var_index.unstack(level =-1)

#flatten table and break multi-level index
eu_df_filter_by_var_flat = pd.DataFrame(eu_df_filter_by_var_index_unstack.to_records())

#rename columns
#isolate variable columns (after our index columns - there should be two remaining)

#start position defined by: number of successful columns - total columns
eu_df_start = len(eu_df_filter_by_var_flat.columns) - euro_variable_present
eu_df_columns_original = eu_df_filter_by_var_flat.columns[eu_df_start:]
eu_df_columns_new_name = []

for column in eu_df_columns_original:
    new_column = column[-9:-2]
    eu_df_columns_new_name.append(new_column)

#create dictionary to rename columns
eu_df_columns_rename_dic = {}
for i in range(0,len(eu_df_columns_new_name)):
    eu_df_columns_rename_dic[eu_df_columns_original[i]] = eu_df_columns_new_name[i]

eu_df_columns_rename_dic

#7 character IDs

eu_df_filter_by_var_flat.rename(columns = eu_df_columns_rename_dic, inplace = True)

eu_df_filter_by_var_flat.head()
```

```python
#===(3.5) SEPARATE COUNTRIES AND CITIES
#function that creates a fork, return table 0 for cities and table 1 for country data
def f_city_country_separate(p_df):
    
    df = p_df.copy()
    
    #create flag for countries
    country_flag = []

    for entry in df.cities:
        if len(entry) == 2:
            country_flag.append("country")
        else:
            country_flag.append("city")

    df["country_flag"] = country_flag
    
    #filter dataframe by flag value and then drop flag column
    df_country = df[df["country_flag"] == "country"].copy()
    df_country.drop(columns = ["country_flag"], inplace = True)
    
    df_city = df[df["country_flag"] == "city"].copy()
    df_city.drop(columns = ["country_flag"], inplace = True)
    
    #return two dataframes
    return df_city,df_country

#run function
eu_df_filter_cities = f_city_country_separate(eu_df_filter_by_var_flat)[0]
eu_df_filter_countries = f_city_country_separate(eu_df_filter_by_var_flat)[1]
```

#### CLUSTER TESTING

Test cluster quality with different techniques:

Code from [here](https://scikit-learn.org/stable/modules/clustering.html#clustering-performance-evaluation)

```python
#===(8.0) CLUSTER TESTING: SET UP

cluster_range = list(range(2,15,1))
cluster_test_list = ["silhouette_score","calinski_harabasz_score","davies_bouldin_score"]

#===(8.1) CLUSTER TESTING: FUNCTION (LINKAGE SET TO WARD)

def f_cluster_test(p_range,p_data,p_tests):
    
    cluster_test_df = pd.DataFrame(p_range, columns = ["Clusters"])
    cluster_test_df.set_index("Clusters", inplace= True)
    #for every type of test do the following
    for test in p_tests:
        f_score_num = []
        f_score_output = []
        
        print(f"{test} processing")
        for number in cluster_range:
            cluster_number = number
            cluster_method_linkage = "ward"

            #CLUSTERING
            cluster = AgglomerativeClustering(n_clusters = cluster_number,affinity = "euclidean", linkage = cluster_method_linkage)
            cluster.fit_predict(p_data)

            #Silhuette score
            model = cluster.fit_predict(p_data)
            labels = cluster.labels_
        
            #run test
            if test == "silhouette_score":
                f_score = sklm.silhouette_score(p_data, labels)
                #force scale the score to help it show up on 1 axis plot
                f_score = f_score * 10000 
            elif test == "calinski_harabasz_score":
                f_score = sklm.calinski_harabasz_score(p_data, labels)
            elif test == "davies_bouldin_score":
                f_score = sklm.davies_bouldin_score(p_data, labels)
                #force scale the score to help it show up on 1 axis plot
                f_score = f_score * 10000
            else:
                f_score = f"Test failed for cluster {number}"
            
            #append score to score list
            f_score_num.append(f_score)
            #readible list
            f_score_output.append(f"Score for {number} clusters is: {f_score}")
        
        cluster_test_df[test] = f_score_num
    
    #return dataframe with number of clusters and results from all tests
    print("cluster test complete")
    return cluster_test_df

#===(8.2) CLUSTER TESTING: CALL FUNCTION

cluster_test_output = f_cluster_test(cluster_range,cluster_data,cluster_test_list)
```

```python
#===(8.3) FIGURE 5

#CLUSTER TESTING: CALINSKI HARABASZ SCORE - higher score is better
#CLUSTER TESTING: SILHOUETTE SCORE        - the closer to positive 1 (10,000 in this version) the better
#CLUSTER TESTING: DAVIES BOULDIN TEST     - lower score is better (also scaled up to go from 10k - 0)

cluster_test_output.columns = ["Silhouette score","Calinski Harabasz score","Davies Bouldin test"]

cluster_test_output.plot(figsize = (20,3))
cluster_test_output.to_csv("1_Data_export/fig_4_cluster_test.csv")
cluster_test_output.head()
```

```python
#===(9.0) CLUSTER ANALYSIS: END TO END HIERARCHICAL CLUSTERING FUNCTION

def f_cluster_end_to_end(p_data,p_cluster_number,p_linkage,p_data_master,p_colour = "Pastel2",p_title = "auto",p_axes = "auto"):
    
    df_return = p_data_master.copy()
    
    for link in p_linkage:
    
        #CLUSTERING
        cluster = AgglomerativeClustering(n_clusters = p_cluster_number,affinity = "euclidean", linkage = link)
        cluster.fit_predict(p_data)
        
        #SCATTER-PLOT TITLE
        plt_title_string = "" 
        column_count = 1
        for column in p_data.columns:
            plt_title_string = plt_title_string + f"{column.upper()}"

            if column_count < len(cluster_data_process.columns):
                plt_title_string = plt_title_string + "_"
                column_count += 1
            else:
                pass
        
        #Title setting conditional
        if p_title == "auto":
            plt_title_string_final = plt_title_string+" CLUSTER PLOT " + link.upper()
        else:
            plt_title_string_final = p_title
            
        ###Add cluster info to DF
        column_name = plt_title_string+"_"+link
        df_return[column_name] = cluster.labels_
        
        #create cluster group names
        cluster_group_list = []
        
        for entry in df_return[column_name]:
            cluster_tag = "Cluster " + str(entry+1)
            cluster_group_list.append(cluster_tag)
        
        df_return["Cluster group"] = cluster_group_list
            
        #Initiative axes, also required to show X axis (https://stackoverflow.com/questions/43578976/pandas-missing-x-tick-labels)
        fig, ax = plt.subplots()
                
        #SCATTER-PLOT TITLE
        f_plot = p_data.plot(
            kind = "scatter",
            title = plt_title_string_final,
            figsize = (20,10),
            
            #column names that will be found in column axis
            x = p_data.columns[0],
            y = p_data.columns[-1],
            
            #xlabel = "X LABEL",
            #ylabel = "Y LAVEL",

            #xlim = (0,500000),
            #ylim = (0,1200000),
            c = cluster.labels_,
            cmap = p_colour,
            alpha = 0.45,
            ax=ax
        )
        
         #axes labels conditional
        if p_axes == "auto":
            pass
        else:
            f_plot.set_xlabel(p_axes[0])
            f_plot.set_ylabel(p_axes[1])
    
        #adds major gridlines
        ax.grid(color='grey', linestyle='-', linewidth=0.25, alpha=0.5)        
        #see colour maps here: https://matplotlib.org/examples/color/colormaps_reference.html
    
    return df_return

    
```
